```python
import pandas as pd
import numpy as np
```

```python
df_admins = pd.read_csv('egz_admins.csv')
df_admins['source_bin'] = df_admins['source_bin'].astype(str).str.zfill(12)
df_admins = df_admins.drop_duplicates(subset=['source_bin'])
```

```python
df_admins = df_admins.rename(columns={'bin': 'egz_cgo_bin'})
df_admins = df_admins.rename(columns={'name_ru': 'egz_cgo_name'})
df_admins = df_admins.rename(columns={'source_bin': 'bin'})
```

```python
df = pd.read_csv("spr_gu_all.csv", sep=';')
```

```python
df['bin'] = df['bin'].astype(str).str.zfill(12)
df = df.drop_duplicates(subset=['bin'])
```

```python
df_final = df.merge(
    df_admins[['bin', 'egz_cgo_bin', 'egz_cgo_name', 'budget_ru']],
    on='bin',
    how='left'
)
```

```python
df_adresses = pd.read_csv('egz_adresses.csv')
```

```python
df_adresses['source_bin'] = df_adresses['source_bin'].astype(str).str.zfill(12)
df_adresses = df_adresses.drop_duplicates(subset=['source_bin'])
df_adresses = df_adresses.rename(columns={'source_bin': 'bin'})
```

```python
df_final = df_final.merge(
    df_adresses[['bin', 'address', 'kato_code']],
    on='bin',
    how='left'
)
```
