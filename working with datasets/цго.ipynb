```python
import pandas as pd
import numpy as np
```

```python
df_spr = pd.read_csv('rgi_spr.csv')
```

```python
df_spr['БИН организации'] = df_spr['БИН организации'].astype(str).str.zfill(12)
```

```python
df_spr['flOguBin'] = df_spr['flOguBin'].astype(str).str.zfill(12)
```

```python
df_spr['flOwnerBin\nБИН Владельца'] = df_spr['flOwnerBin\nБИН Владельца'].astype(str).str.zfill(12)
```

```python
df_spr_new = df_spr[df_spr["flOwnerBin\nБИН Владельца"]==df_spr['flOguBin']]
```

```python
df_spr_new = df_spr_new.rename(columns={'БИН организации': 'bin'})
```

```python
df_egz = pd.read_csv("df_egz.csv")
```

```python
df_egz['bin'] = df_egz['bin'].astype(object) 
```

```python
# Преобразуем 'bin' к строковому типу без '.0', пробелов и символов переноса
df_egz['bin'] = df_egz['bin'].astype(str).str.strip().str.replace('.0', '', regex=False)
df_spr_new['bin'] = df_spr_new['bin'].astype(str).str.strip().str.replace('.0', '', regex=False)
```

```python
df_final = df_egz.merge(
    df_spr_new[['bin', 'Наименование организации', 'flOwnerBin\nБИН Владельца']],
    on='bin',
    how='left'
)
```

```python
df_final['bin'] = df_final['bin'].astype(str)
df_spr_new['bin'] = df_spr_new['bin'].astype(str)
```

```python
df_final['bin'] = df_final['bin'].astype(str).str.zfill(12)
df_final['egz_cgo_bin'] = df_final['egz_cgo_bin'].astype(str).str.replace(r'\.0$', '', regex=True)
df_final['egz_cgo_bin'] = df_final['egz_cgo_bin'].astype(str).str.zfill(12)
```

```python
df_final['kato_code'] = df_final['kato_code'].astype(str).str.replace(r'\.0$', '', regex=True)
df_final['kato_code'] = df_final['kato_code'].astype(str).str.zfill(9)
df_final['egz_cgo_bin'] = df_final['egz_cgo_bin'].astype(str).str.replace(r'\.0$', '', regex=True)
df_final['egz_cgo_bin'] = df_final['egz_cgo_bin'].astype(str).str.zfill(12)
```

```python
df_final['kato_code'] = df_final['kato_code'].astype(str).str.zfill(9)
```

```python
df_final['OwnerBin'] = df_final['OwnerBin'].astype(str).str.zfill(12)
```

```python
df_final['OwnerBin'] = df_final['OwnerBin'].astype(str).str.replace(r'\.0$', '', regex=True)
```

```python
df_final = pd.read_csv('df_final.csv')
```

```python
df_final = df_final.rename(columns={'flOwnerBin\nБИН Владельца': 'OwnerBin'})
df_final = df_final.rename(columns={'Наименование организации': 'rgi_name'})
```
